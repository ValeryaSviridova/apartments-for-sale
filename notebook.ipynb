–î—É–±–ª–∏–∫–∞—Ç–æ–≤ –º–Ω–æ–≥–æ, —Å–∞–º—ã–π –ø—Ä–æ—Å—Ç–æ–π –∏ –±—ã—Å—Ç—Ä—ã–π –≤–∞—Ä–∏–∞–Ω—Ç —É–±—Ä–∞—Ç—å —Ç–∏–ø—ã –Ω–∞—Å–µ–ª–µ–Ω–Ω—ã—Ö –ø—É–Ω–∫—Ç–æ–≤ –∏ –æ—Å—Ç–∞–∏–≤—Ç—å —Ç–æ–ª—å–∫–æ –Ω–∞–∑–≤–∞–Ω–∏—è —Å –ø–æ–º–æ—â—å—é split, –æ—Å—Ç–∞–≤–∏–≤ —Ç–æ–ª—å–∫–æ –ø–æ—Å–ª–µ–¥–Ω–µ–µ —Å–ª–æ–≤–æ –≤ —Å—Ç–æ–ª–±—Ü–µ.

–û—Ç–ª–∏—á–Ω—ã–π –∫–æ–¥

–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π —Å–æ–∫—Ä–∞—Ç–∏–ª–æ—Å—å –Ω–∞ 1/4. –î–∞–Ω–Ω—ã–µ –æ –±–ª–∏–∂–∞–π—à–∏—Ö –ø–∞—Ä–∫–∞—Ö –∏ –ø—Ä—É–¥–∞—Ö –º–æ–∂–Ω–æ –Ω–µ –æ–±—Ä–∞–±–∞—Ç—ã–≤–∞—Ç—å, —Ç–∞–∫ –∫–∞–∫ –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞ –Ω–∞–º –Ω–µ –Ω—É–∂–Ω—ã —ç—Ç–∏ —Å—Ç–æ–ª–±—Ü—ã, –Ω–æ –º–æ–∂–µ–º –∑–∞–º–µ–Ω–∏—Ç—å –∏—Ö –º–µ–¥–∏–∞–Ω–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏, –ø–æ —Ç–∏–ø—É –Ω–∞—Å–µ–ª–µ–Ω–Ω–æ–≥–æ –ø—É–Ω–∫—Ç–∞.

```python
#—Å–æ–∑–¥–∞–µ–º –¥–≤–∞ —Ü–∏–∫–ª–∞, –∫–æ—Ç–æ—Ä—ã–µ —Å—á–∏—Ç–∞—é—Ç –º–µ–¥–∏–∞–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ parks_nearest –∏ ponds_nearest –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç 
#–Ω–∞–∑–≤–∞–Ω–∏—è –Ω–∞—Å–µ–ª–µ–Ω–Ω–æ–≥–æ –ø—É–Ω–∫—Ç–∞

for t in data['local'].unique():
    data.loc[(data['local'] == t) & (data['parks_nearest'].isna()), 'parks_nearest'] = \
    data.loc[(data['local'] == t), 'parks_nearest'].median()

for t in data['local'].unique():
    data.loc[(data['local'] == t) & (data['ponds_nearest'].isna()), 'ponds_nearest'] = \
    data.loc[(data['local'] == t), 'ponds_nearest'].median()     
```

```python
#—Å –ø–æ–º–æ—â—å—é –ª–æ–≥–∏—á–µ—Å–∫–æ–π –∏–Ω–¥–µ–∫—Å–∞—Ü–∏–∏ –≤—ã–±–∏—Ä–∞–µ–º —Å—Ç—Ä–æ–∫–∏ —Å –ø—Ä–æ–ø—É—Å–∫–∞–º–∏ –∏ –∑–∞–ø–æ–ª–Ω—è–µ–º –∏—Ö —Ä–∞–∑–Ω–∏—Ü–µ–π
data.loc[data['kitchen_area'].isna(), 'kitchen_area'] = data.loc[data['kitchen_area'].isna(), 'kitchen_area'] - data.loc[data['kitchen_area'].isna(), 'living_area']
data.loc[data['living_area'].isna(), 'living_area'] = data.loc[data['living_area'].isna(), 'total_area'] - data.loc[data['living_area'].isna(), 'kitchen_area']
```

```python
#—Ü–∏–∫–ª –ø—Ä–æ—Ö–æ–¥–∏—Ç –ø–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏—è–º –∫–æ–ª-–≤–∞ –∫–æ–º–Ω–∞—Ç, —Å—á–∏—Ç–∞–µ—Ç –º–µ–¥–∏–∞–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∫—É—Ö–Ω–∏ –∏ –∑–∞–ø–æ–ª–Ω—è–µ—Ç –∏–º –ø—Ä–æ–ø—É—Å–∫–∏
for t in data['rooms'].unique():
    data.loc[(data['rooms'] == t) & (data['kitchen_area'].isna()), 'kitchen_area'] = \
    data.loc[(data['rooms'] == t), 'kitchen_area'].median()  

for t in data['rooms'].unique():
    data.loc[(data['rooms'] == t) & (data['living_area'].isna()), 'living_area'] = \
    data.loc[(data['rooms'] == t), 'living_area'].median()    
```

    def quartile_range(data, column):
        q1 = data[column].quantile(0.25)
        q3 = data[column].quantile(0.75)
        iqr = q3 - q1
        dfq = data.loc[(data[column] < q3 + 1.5*iqr) & (data[column] > q1 - 1.5*iqr), column]
        return dfq

    list_features = ['last_price', 'total_area', 'kitchen_area', 'living_area', 'days_exposition', 'ceiling_height']
